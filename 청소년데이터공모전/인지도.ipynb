```python
# 운영 체제마다 한글이 보이게 하는 설정
# 윈도우
import platform
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

# '-' 기호 보이게 하기
plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Windows':
    path = "C:\Windows\Fonts\malgun.ttf"
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
```

```python
knowledge_data = pd.read_csv('./인지도.csv', encoding='ANSI')
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Get unique categories in '특성별(1)'
categories = knowledge_data['특성별(1)'].unique()

# Iterate over each unique category in '특성별(1)' and create bar charts for '알고있음' and '모름'
for category in categories:
    # Filter data for the current category
    category_data = knowledge_data[knowledge_data['특성별(1)'] == category]
    
    # Define the x locations for the groups
    x = np.arange(len(category_data))
    
    # Define the width of the bars
    width = 0.35
    
    # Plotting the bar chart
    plt.figure(figsize=(10, 6))
    plt.bar(x - width/2, category_data['알고있음'], width, label='알고있음', color='skyblue')
    plt.bar(x + width/2, category_data['모름'], width, label='모름', color='salmon')
    
    # Customize the plot
    plt.xticks(x, category_data['특성별(2)'], rotation=45, ha="right")
    plt.ylabel("비율 (%)")
    plt.title(f"{category}별 인지도 (알고있음/모름)")
    plt.legend()
    plt.tight_layout()
    plt.show()
```
