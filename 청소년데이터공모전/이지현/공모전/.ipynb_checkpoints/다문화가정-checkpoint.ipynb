```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import koreanize_matplotlib
import warnings
import os

%config InlineBackend.figure_format='retina'
warnings.filterwarnings('ignore')
os.environ['KMP_DUPLICATE_LIB_OK']='True'
```

```python
data7 = pd.read_csv("다문화가정_인지율.csv", encoding='euc-kr', header=1, index_col=0)
```

```python
temp_index = data7.copy()
temp_index.reset_index(drop= False,inplace=True)
temp_index
```

```python
# 전체 수 비교
plt.figure(figsize=(4,4))
bar_width = 0.25

plt.title("전체 인지와 이용률 차이")
bar1 = plt.bar(range(3), height=data7.iloc[0,[1,2,4]].astype(int), width=bar_width)

for bar in bar1:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 1), ha='center', va='bottom',fontsize=7)



plt.xticks(range(3), labels=['다문화 가정 전체','인지하는 가정', '이용 가정'])  # x축 레이블 위치 조정
plt.tick_params(axis='x', direction='in', length=3, pad=8, labelsize=8, top=True)
plt.grid(axis='y', color='darkgreen', alpha=0.6, linestyle='--')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 이용자 이용연령에 따른 그래프  
plt.figure(figsize=(4,4))
bar_width = 0.25

plt.title("연령에 따른 이용자 수 비교")

# x축 위치 설정
x = np.arange(len(data7.iloc[6:10, 0]))  # x축 위치를 숫자로 설정

# '이용' 데이터 바 그래프
bar1 = plt.bar(x, height=data7.iloc[6:10, 2].astype(float), width=bar_width, label='연령별 인지 비율(%)')

# '이용안함' 데이터 바 그래프 (x축을 바 너비만큼 이동)
bar2 = plt.bar(x + bar_width, height=data7.iloc[6:10, 4].astype(float), width=bar_width, label='실제 이용하는 비율(%)')

# 각 막대 위에 값 표시
for bar in bar1:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 1), ha='center', va='bottom',fontsize=7)

for bar in bar2:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 1), ha='center', va='bottom',fontsize=7)

# 축 조정
plt.xticks(x + bar_width / 2, labels=['9∼11세', '12∼14세', '15∼17세', '18세 이상'])  # x축 레이블 위치 조정
plt.tick_params(axis='x', direction='in', length=3, pad=4, labelsize=10, top=True)

plt.grid(axis='y', color='darkgreen', alpha=0.6, linestyle='--')
plt.legend()
plt.show()
```
