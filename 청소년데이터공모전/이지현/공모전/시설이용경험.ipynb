```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import koreanize_matplotlib
import warnings
import os

%config InlineBackend.figure_format='retina'
warnings.filterwarnings('ignore')
os.environ['KMP_DUPLICATE_LIB_OK']='True'
```

```python
data3 = pd.read_csv("시설이용경험.csv", encoding='euc-kr', header=1, index_col=0)
```

```python
temp_index = data3.copy()
temp_index.reset_index(drop= False,inplace=True)
temp_index
```

```python
# 성별에 따른 이용비교 그래프  -> 근데 이 이용자가 
plt.figure(figsize=(4, 4))
bar_width = 0.25

plt.title("성별에 따른 이용자 수 비교")

# x축 위치 설정
x = np.arange(len(data3.iloc[:3, 0]))  # x축 위치를 숫자로 설정

# '이용' 데이터 바 그래프
bar1 = plt.bar(x, height=data3.iloc[:3, 1].astype(float), width=bar_width, label='이용')

# '이용안함' 데이터 바 그래프 (x축을 바 너비만큼 이동)
bar2 = plt.bar(x + bar_width, height=data3.iloc[:3, 2].astype(float), width=bar_width, label='이용안함')

for bar in bar1:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 1), ha='center', va='bottom',fontsize=7)

for bar in bar2:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 1), ha='center', va='bottom',fontsize=7)

# 축 조정
plt.xticks(x + bar_width / 2, labels=['전체', '남성', '여성'])  # x축 레이블 위치 조정
plt.tick_params(axis='x', direction='in', length=3, pad=4, labelsize=12, top=True)

plt.grid(axis='y', color='darkgreen', alpha=0.6, linestyle='--')
plt.legend()
plt.show()
```
