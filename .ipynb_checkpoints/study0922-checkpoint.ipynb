## Inner Join
Inner Join은 두 테이블간의 공통된 값을 가지는 행만을 반환 합니다.

공통된 값을 가지지 않는 행들은 결과에 포함되지 않습니다.




## Outer Join
Outer Join은 Inner Join과 달리, 조인 조건을 만족하지 않는 행도 결과에 포함시킵니다.

Outer Join은 Left Outer Join, Right Outer Join, Full Outer Join으로 나뉩니다.

### Left Outer Join
왼쪽 테이블의 모든 행을 포함하고, 오른쪽 테이블과 일치하는 값이 없는 경우 NULL을 반환합니다.



### Right Outer Join
오른쪽 테이블의 모든 행을 포함하고, 왼쪽 테이블과 일치하는 값이 없는 경우 NULL을 반환합니다.



### Full Outer Join
양쪽 테이블의 모든 행을 포함하고, 일치하는 값이 없는 경우에는 NULL을 반환합니다.




- Inner Join:	양쪽 데이터프레임에 모두 존재하는 키의 행만 포함

- Left Join:	왼쪽 데이터프레임의 모든 행을 포함하고, 오른쪽에서 일치하는 데이터가 있으면 병합

- Right Join:	오른쪽 데이터프레임의 모든 행을 포함하고, 왼쪽에서 일치하는 데이터가 있으면 병합

- Full Outer Join:	양쪽 데이터프레임의 모든 행을 포함하고, 일치하지 않는 데이터는 NaN 처리

```python
import pandas as pd

df1 = pd.DataFrame({
    '고객ID': [1, 2, 3],
    '이름': ['김철수', '이영희', '박민수']
})

df2 = pd.DataFrame({
    'ID': [1, 2, 4],
    '구매금액': [10000, 20000, 15000]
})

# 키 열 이름이 다를 때
merged_df = pd.merge(df1, df2, left_on='고객ID', right_on='ID', how='outer')
print('키 열 이름이 다를 때의 Full Outer Join 결과:')
merged_df
```

```python
df1 = pd.DataFrame({
    '키1': ['A', 'B', 'C'],
    '키2': [1, 2, 3],
    '값1': [10, 20, 30]
})

df2 = pd.DataFrame({
    '키1': ['A', 'B', 'D'],
    '키2': [1, 2, 4],
    '값2': [100, 200, 300]
})

# 여러 키로 병합
multi_key_merge = pd.merge(df1, df2, on=['키1', '키2'], how='outer')
print('여러 키로 병합한 결과:')
multi_key_merge
```

# pivot
Pivot은 데이터를 재구조화하여 특정 열의 값들을 새로운 열로 변환합니다.

이를 통해 긴(long) 형식의 데이터를 넓은(wide) 형식으로 바꿀 수 있습니다.

- 긴 형식(Long Format): 데이터가 관측값별로 행에 길게 나열된 형태.

- 넓은 형식(Wide Format): 변수들이 열로 분산되어 넓게 펼쳐진 형태.


### DataFrame.pivot(index={ }, columns={ }, values={ })
- index: 새로운 데이터프레임에서 행 인덱스로 사용할 열 이름.
- columns: 새로운 열 이름으로 사용할 열 이름.
- values: 피벗된 데이터의 값으로 사용할 열 이름.

```python
import pandas as pd

data = {
    '날짜': ['2021-01-01', '2021-01-01', '2021-01-02', '2021-01-02'],
    '지역': ['서울', '부산', '서울', '부산'],
    '확진자수': [100, 50, 80, 60]
}

df = pd.DataFrame(data)
df
```

```python
pivot_df = df.pivot(index='날짜', columns='지역', values='확진자수')
pivot_df
```

```python
data = {
    '날짜': ['2021-01-01', '2021-01-01', '2021-01-02', '2021-01-02'],
    '지역': ['서울', '서울', '부산', '부산'],
    '확진자수': [100, 80, 50, 60]
}

df = pd.DataFrame(data)
df
```

```python
pivot_table_df = pd.pivot_table(df, values='확진자수', index='날짜', columns='지역', aggfunc='sum')
pivot_table_df.index.name = None
pivot_table_df
```

# Melt
Melt는 여러 열로 분산된 데이터를 하나의 열로 모아 긴(long) 형식으로 변환합니다. 

이는 피벗의 반대 과정이라고 볼 수 있습니다.

여러 변수에 해당하는 열들을 하나의 열로 모아 분석하기 쉽게 만듭니다.

### pd.melt(frame, id_vars={}, value_vars={ }, var_name=' ', value_name=' ')

- frame: 변환할 데이터프레임
- id_vars: 식별자 변수로 남길 열 이름 리스트
- value_vars: 행으로 변환할 열 이름 리스트
- var_name: 새로운 'variable' 열의 이름 (기본값은 'variable')
- value_name: 새로운 'value' 열의 이름 (기본값은 'value')

```python
data = {
    '지역': ['서울', '서울', '부산', '부산'],
    '연도': [2020, 2021, 2020, 2021],
    'Q1': [100, 150, 80, 90],
    'Q2': [120, 160, 85, 95]
}

df = pd.DataFrame(data)
df
```

```python
melted_df = pd.melt(df, id_vars=['지역', '연도'], var_name='분기', value_name='매출액')
melted_df
```

8.4 참고 질문

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 삼성전자 주식 6개월 데이터
df = pd.read_csv('./samsung.csv')

df
```

```python
df.info()
```

```python
# 날짜 정렬
df = df.sort_values('날짜')
df = df.reset_index(drop=True)
```

```python
df['DateTime'] = pd.to_datetime(df['날짜'])

# Year, Month, Day, Hour 분리
df['Year'] = df['DateTime'].dt.year
df['Month'] = df['DateTime'].dt.month
df['Day'] = df['DateTime'].dt.day
df['Open'] = df['시가'].str.replace(',', '').astype(int)
df['Close'] = df['종가'].str.replace(',', '').astype(int)
df['High'] = df['고가'].str.replace(',', '').astype(int)
df['Low'] = df['저가'].str.replace(',', '').astype(int)
# 데이터 선택
df = df.loc[:, ['DateTime', 'Year', 'Month','Day', 'Open', 'Close', 'High', 'Low']]
df
```

```python
# 최근 3일간 Open 열 평균
df['Open_mean_3'] = round(df['Open'].rolling(window=3, min_periods=1).mean(), 1)
df['Open_mean_7'] = round(df['Open'].rolling(window=7, min_periods=1).mean(), 1)
df
```

```python
# OCHL 컬럼 생성
df['OCHL_mean'] = (df['Open'] + df['Close'] + df['High'] + df['Low']) / 4
df
```

```python
# 시각화
plt.plot(df['DateTime'], df['OCHL_mean'])
plt.show()
```

```python
# 피벗
df_pivot = df.pivot(index=['Year', 'Month'], columns='Day', values='OCHL_mean')

df_pivot.columns.name = None

df_pivot.reset_index(drop=False, inplace=True)

df_pivot
```

```python
# 결측치 확인
df_pivot.isna().sum()
```

```python
# 결측치 채우기
df_pivot_fill = df_pivot.ffill(axis=1)
df_pivot_fill
```

```python
# 결측치 확인
df_pivot_fill.isna().sum()
```

```python
# 언피벗
df_melt= pd.melt(df_pivot, 
                      id_vars=['Year', 'Month'],
                      # value_vars=range(1, 32),
                      var_name='Day',
                      value_name='OCHL_mean')

# 정렬
df_melt.sort_values(by=['Year', 'Month', 'Day'], ascending=True, inplace=True)

# 결측치 제거(31일이 꽉 차지 않는 월에 결측치 존재)
df_melt.dropna(inplace=True)

# 인덱스 초기화
df_melt.reset_index(drop=True, inplace=True)

# 데이터 형식 변경
df_melt['OCHL_mean'] = df_melt['OCHL_mean'].astype(int)

# 확인
df_melt
```

```python
# Date 컬럼 생성
df_melt['Date'] = pd.to_datetime(df_melt[['Year', 'Month', 'Day']])
df_melt
```

```python
# 컬럼 정리
df_melt = df_melt[['Date', 'OCHL_mean']]
df_melt
```

```python
# 시각화
plt.plot(df_melt['Date'], df_melt['OCHL_mean'])
plt.show()
```
